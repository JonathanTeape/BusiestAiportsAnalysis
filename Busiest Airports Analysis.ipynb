## Importing Necessary Libraries

```python
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline

```

## Reading in Data

```python
df =pd.read_csv ('airports.csv')
```

 ## Exploring Data

```python
df.head ()
```

```python
missing_values_count = df.isnull().sum() ##checking for missing values 
missing_values_count
```

```python
df = df.drop ("Code", axis=1) ##Removing country code as it has no revelance to data 
df
```

```python
year=df.groupby("Year").sum()['Passengers'].sort_values(ascending=False) ##Sorts passengers by Year 
```

```python
airport=df.groupby('Airport').sum()['Passengers'].sort_values (ascending=False) #Shows which aiports are the most crowded 
```

```python
c_df = df.groupby('Country').sum()['Passengers'].sort_values (ascending=False) #Shows which countries are the most crowded 
c_df
```

```python
##Shows top 5 countries which have the most busiest aiports.
c_df[:5].plot(kind="bar",figsize=(10,10),color = ("orange"))
plt.xticks(rotation=0)
plt.xlabel("Country",fontsize=(20))
plt.ylabel("Passenger Count in Billions",fontsize=(20))
plt.title("Most Crowded Airport",fontsize=(20));
```

```python
year.plot(kind="bar",figsize=(10,10),color = ("orange"))
plt.xticks(rotation=0)
plt.xlabel("Country",fontsize=(20))
plt.ylabel("Passenger Count in Billions",fontsize=(20))
plt.title("Most Busiest Years",fontsize=(20));
```

```python

explode = (0.08, 0.00, 0.00, 0.00, 0.00)
airport[:5].plot(kind="pie",figsize=(10,10), autopct='%1.2f%%', shadow=True,startangle=90, explode=explode,textprops={'fontsize':15},
)
plt.title("Top 5 Busiest Airports",fontsize=(20))
plt.ylabel (None)
plt.legend(loc="center", bbox_to_anchor=(1.5,1));
```

## Conclusion


After exploring the dataset, it is evident that the busiest airports are American, China, Japan, and India. In addition, the busiest airports by passengers are Hartsfield-Jackson Atlanta International Airport and Being Beijing Capital International Airport. Furthermore, the busiest year from the dataset is 2019. 2020 was the lowest amount of passengers which can most likely be attributed to covid.
